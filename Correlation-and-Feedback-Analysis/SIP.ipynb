```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df= pd.read_csv('Unified002.csv')
```

# Data Cleaning

```python
# Replace "yes" with "Yes" in the entire DataFrame
df = df.replace('yes', 'Yes')
```

```python
# copy last alfabate of Ref Id and past it in new column Gender

df['Gender'] = df['Ref Id'].str[-1]
```

# EDA

```python

# Group the data
gender_completion = df.groupby(['Gender', 'Complete or Not']).size().unstack(fill_value=0)

# Create the bar plot
ax = gender_completion.plot(kind='bar', figsize=(8, 6))

# Add counts on top of the bars
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Completion Status by Gender')
plt.xticks(rotation=0)  # X axis labale rotaion
plt.legend(title='Complete or Not')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define age bins and corresponding labels
bins = [10, 20, 25, 30, 35, float('inf')]
labels = ['10-19', '20-24', '25-29', '30-34', '35+']

# Assign age groups to the 'Age_Group' column
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Calculate the normalized frequency of 'Level' for each 'Age_Group'
correlation_matrix = (
    df.groupby('Age_Group')['Level']
    .value_counts(normalize=True)
    .unstack(fill_value=0)  # Handle missing levels with 0
)

# Drop the 4th column from the correlation_matrix
correlation_matrix = correlation_matrix.drop(correlation_matrix.columns[3], axis=1)

# Create a heatmap to visualize the correlation
plt.figure(figsize=(12, 8))  # Increased size for better readability
sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='viridis',
    fmt=".2%",
    cbar_kws={'label': 'Proportion'}  # Add color bar label
)

# Add plot titles and axis labels
plt.title('Proportion of Levels by Age Group', fontsize=16)
plt.xlabel('Level', fontsize=12)
plt.ylabel('Age Group', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()  # Adjust layout to avoid overlap
plt.show()
```

```python
# Define the columns for correlation analysis
columns_to_correlate = [
    'Did this course meet your expectations?',
    'Did this course help you improve your skills?',
    'Did you gain practical knowledge from this course?',
    'Did the course content effectively cover the topics it promised?',
    'Were the materials and resources provided sufficient?',
    'Would you recommend this course to others?',
    'Did the course length feel appropriate for the content covered?',
    'Did you experience any technical issues during the course?',
    'Do you believe the course provided good value for the cost?'
]

# Convert the relevant columns to numerical representation for correlation
# Handle other responses or missing values as appropriate for your data.

for col in columns_to_correlate:
    df[col] = df[col].map({'Yes': 1, 'No': 0})


# Calculate the correlation between 'Course Name' and other columns
correlation_results = {}

for col in columns_to_correlate:
  try:
    correlation = df.groupby('Course Name')[col].mean()
    correlation_results[col] = correlation
  except Exception as e:
    print(f"Error calculating correlation for {col}: {e}")

# Print or further process the correlation results
for col, correlation in correlation_results.items():
    print(f"Correlation between Course Name and '{col}':\n{correlation}\n")
```
